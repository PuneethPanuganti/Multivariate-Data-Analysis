```python
#Count missing values for each column of the dataframe df
facebook_df.isnull().sum()
```

```python
#The code below drops all missing rows with missing values.
#df.dropna() | Drop all rows that contain null values
facebook_df.dropna(inplace=True)
facebook_df.isnull().sum()
```

```python
y = facebook_df["like"]
X = facebook_df.drop(["like"], axis=1)

```

```python
#An intercept is not included by default so it  should be added by the user.
X = sm.add_constant(X)
full_model = sm.OLS(y, X).fit()
full_model.summary()
```

# Q2 Bootstrap confidence interval

```python
facebook_df.describe()
```

Your tasks is to use the boostrap method to construct a 95% CI for the coefficients for **comment** and **share** using the same model as in Q1. Do you get similar CI to Q1? Why or why not?

```python
from tqdm import tqdm
import statsmodels.formula.api as smf
```

```python
bootstrapped_ests = []
for i in tqdm(range(5000)):
    facebook_like_data_boot = facebook_df.sample(n=int(len(facebook_df) * 0.8), replace=True) 
    facebook_like_model_boot = smf.ols(formula='like ~ comment + share', data=facebook_like_data_boot).fit()
    bootstrapped_ests.append(facebook_like_model_boot.params[['comment', 'share']])
    
b_comments, b_shares = zip(*bootstrapped_ests)

print("The 95% bootstrapped CI of b_comments is [{:.2f}, {:.2f}].".format(
    np.percentile(b_comments, 2.5),
    np.percentile(b_comments, 97.5)))

print("The 95% bootstrapped CI of b_shares is [{:.2f}, {:.2f}].".format(
    np.percentile(b_shares, 2.5),
    np.percentile(b_shares, 97.5)))
```

## Q4 - Compare out-of-sample perormance

- Split the data to 90% train and 10% test set. 
- Train three models using the train set:
   * `LassoLarsIC` model (a model that uses BIC to select variables)
   * a full model (a model that includes all X variables)
   * a `LassoCV` model (a model that uses CV on train set to select variables)
- Remember to train a scaler using train set only, and transform both X_train and X_test. 
- Compare their test-set performance using MSE. Which model has the best performance? 

```python

from scipy import interpolate
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import linear_model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state = 4)
X_train.shape
X_test.shape
scaler = preprocessing.StandardScaler().fit(X_train)
model_lassoCV = linear_model.LassoCV(cv=5, normalize=False) # note we don't need to normalize again
model_lassoCV.fit(scaler.transform(X_train), y_train) # note we transform X_train

#predicting the test set and seeing the performance
y_hat_test = model_lassoCV.predict(scaler.transform((X_test))) 


from sklearn import metrics
metrics.mean_squared_error(y_true=y_test, y_pred=y_hat_test)
```

```python
#Lasso Lars IC
model_lassolars = linear_model.LassoLarsIC(criterion='bic')
model_lassolars.fit(scaler.transform(X_train), y_train)

#predict test set and see performance
y_hat_test_1 = model_lassolars.predict(scaler.transform((X_test))) 

#seeing the metric
metrics.mean_squared_error(y_true=y_test, y_pred=y_hat_test_1)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

#Full Model
regr = linear_model.LinearRegression(fit_intercept=True)
regr.fit(scaler.transform(X_train), y_train)
y_pred1 = regr.predict(scaler.transform(X_test))

#Mean squared error
print('Mean squared error: %.2f'
     % mean_squared_error(y_test, y_pred1))
```
